```python
import random
```

```python
def gerar_numeros_mega_sena():
    
    num_gerado = list()
    num_max = 6
    contagem = 0

    while contagem < num_max:
        
        r = random.randint(1,10)
        
        if r not in num_gerado:
            contagem += 1
            num_gerado.append(r)
        
    num_gerado.sort()
    print(f'Os numeros sorteados foram: {num_gerado}')
    return num_gerado
```

```python
def computador_joga_mega_sena():
    num_gerado = list()
    num_max = 6
    contagem = 0

    while contagem < num_max:
        
        r = random.randint(1,10)
        
        if r not in num_gerado:
            contagem += 1
            num_gerado.append(r)
        
    num_gerado.sort()
    print(f'O primeiro numero que o pc sorteou foi: {num_gerado}')
    return num_gerado
```

```python
def qtd_jogos_precisos(primeiro_jogo, fezinha_pc):
    qtd_jogos_para_ganhar = 1
    contagem = 0
    num_max = 6

    while primeiro_jogo != fezinha_pc:
        qtd_jogos_para_ganhar += 1
        print(fezinha_pc)
        fezinha_pc.clear()
        contagem = 0
        while contagem < num_max:
            r = random.randint(1,10)
            if r not in fezinha_pc:
                contagem += 1
                fezinha_pc.append(r)
                fezinha_pc.sort()

    print(fezinha_pc)
    print(qtd_jogos_para_ganhar)
```

```python
# Sorteia os numeros do jogo 001
primeiro_jogo = gerar_numeros_mega_sena()

# Sorteia os numeros do jogo 001
fezinha_pc = computador_joga_mega_sena()

# Verifica quantos jogos sao precisos pra acertar
qtd_jogos_precisos(primeiro_jogo, fezinha_pc)
```
